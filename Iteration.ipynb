And line 16 can be read as: The array called "ans_1", add the value sqrt(the second to last/previous value of ans_1) to your list.

Since i >= 1 now, we have multiple values in ans_1, but this iterative process works by using the previous value in the list. (And you have to specify a value becuase we aren't squarerooting the entire list.) I don't want to (and can't) use ans_1[i], because I don't have it yet. I'm making it right now!

But in line 17, I can use ans_1[i] because I have just made it and added it to the list. So that is why I can plot it. I use the i values in the x position, and the ans_1 values in the y position.

```python
# f1(x)
import math
import numpy as np
import matplotlib.pyplot as plt

seed_1 = 5
iterations_1 = 10

ans_1 = []
for i in range(0, iterations_1):
  if i == 0:
    ans_1.append(math.sqrt(seed_1))
    plt.scatter(i, seed_1)

  if i >= 1:
    ans_1.append(math.sqrt(ans_1[i-1]))
    plt.scatter(i, ans_1[i])
```

```python
# f2(x)
import math
import numpy as np
import matplotlib.pyplot as plt

seed_2 = 5
iterations_2 = 10

ans_2 = []
for i in range(0, iterations_2):
  if i == 0:
    ans_2.append(math.sqrt(math.sqrt(seed_2)))
    plt.scatter(i, seed_2)

  if i >= 1:
    ans_2.append(math.sqrt(math.sqrt(ans_2[i-1])))
    plt.scatter(i, ans_2[i])
```

```python
# f3(x)
import math
import numpy as np
import matplotlib.pyplot as plt

seed_3 = 5
iterations_3 = 10

ans_3 = []
for i in range(0, iterations_3):
  if i == 0:
    ans_3.append(math.sqrt(math.sqrt(math.sqrt(seed_3))))
    plt.scatter(i, seed_3)

  if i >= 1:
    ans_3.append(math.sqrt(math.sqrt(math.sqrt(ans_3[i-1]))))
    plt.scatter(i, ans_3[i])
```
